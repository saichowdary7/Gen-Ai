```python
from transformers import pipeline
```

```python
llm_pipeline = pipeline(task='sentiment-analysis', model='tabularisai/multilingual-sentiment-analysis') # Changed 'sentiment Analysis' to 'sentiment-analysis'
llm_pipeline('I love you')[0]['label']
```

```python
def sentiment_analysis(text):
    return llm_pipeline(text)[0]['label']
```

```python
import gradio as gr
demo = gr.Interface(fn=sentiment_analysis, inputs='text', outputs='text')
demo.launch()
```
